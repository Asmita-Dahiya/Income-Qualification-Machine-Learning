```python
#Predict the accuracy using random forest classifier and Check the accuracy using random forest with cross validation.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_data=train_data.drop('Target',axis=1)
Y_data=train_data.Target
```

```python
X_data_col=X_data.columns
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
X_data_1=SS.fit_transform(X_data)
X_data_1=pd.DataFrame(X_data_1,columns=X_data_col)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_data_1,Y_data,test_size=0.25,stratify=Y_data,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier()
```

```python
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_predict,Y_test)
```

```python
print(accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_predict,Y_test))
```

```python
print(confusion_matrix(y_predict,Y_test))
```
